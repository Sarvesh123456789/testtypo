```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch the webpage content
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Locate the correct table using CSS selectors
table = soup.select_one("#style-1 div + div .historical_data_table")

# Extract data into a DataFrame
tesla_revenue = pd.DataFrame([
    {"Date": row.select_one("td:nth-of-type(1)").text.strip(),
     "Revenue": row.select_one("td:nth-of-type(2)").text.strip()}
    for row in table.select("tr")[1:]  # Skip header row
])

# Display the last five rows
print(tesla_revenue.tail())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch the webpage content
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {"User-Agent": "Mozilla/5.0"}  # Mimic a real browser request
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Locate the correct revenue table
table = soup.find("table", {"class": "historical_data_table"})

# Extract data into a DataFrame
tesla_revenue_data = []
for row in table.find_all("tr")[1:]:  # Skip header row
    columns = row.find_all("td")
    date = columns[0].text.strip()  # First column = Date
    revenue = columns[1].text.strip()  # Second column = Revenue
    tesla_revenue_data.append({"Date": date, "Revenue": revenue})

# Convert to DataFrame
tesla_revenue = pd.DataFrame(tesla_revenue_data)

# Display the last five rows
print(tesla_revenue.tail())
```

```python
import matplotlib.pyplot as plt
import yfinance as yf

# Download Tesla stock data
tesla_data = yf.Ticker("TSLA").history(period="max")

# Define the make_graph function
def make_graph(data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data.index, data["Close"], label="Tesla Stock Price", color='blue')
    plt.xlabel("Date")
    plt.ylabel("Stock Price (USD)")
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()

# Call the function with Tesla data and graph title
make_graph(tesla_data, "Tesla Stock Price Over Time")
```
